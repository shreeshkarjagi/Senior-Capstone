```python
import matplotlib.pyplot as plt

# x-values
x = [0, 2.33, 4.5, 6, 8.167, 10.167, 11.5]

# y-values
textile_y = [0, 0.77, 1.53, 2.1, 2.87, 3.6, 4.22]
meg_fiber_y = [0, 0.67, 1.4, 1.93, 2.7, 3.38, 3.96]
thin_film_y = [0, 0.05, 0.1, 0.1341, 0.19, 0.23, 0.27]

# create figure and axis objects
fig, ax = plt.subplots()

# plot each line
ax.plot(x, textile_y, label='Textile (100% cotton)')
ax.plot(x, meg_fiber_y, label='MEG Fiber')
ax.plot(x, thin_film_y, label='Thin Film')

# set axis labels and title
ax.set_xlabel('Time (h)', fontsize=14)
ax.set_ylabel('Water vapor transmission (mL)', fontsize=14)
ax.set_title('Water Vapor Transmission of Different Materials', fontsize=14)

# set legend font size
ax.legend(fontsize=14)

# set transparent background
fig.set_facecolor('w')
fig.patch.set_alpha(0)

# show plot
plt.show()
```
